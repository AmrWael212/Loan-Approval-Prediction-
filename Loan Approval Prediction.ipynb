```python
numeric_cols = df.select_dtypes(include='number').columns
outlier_indices = set()
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
    outlier_indices.update(outliers)
df_clean = df.drop(index=outlier_indices).reset_index(drop=True)
```

```python
edu=pd.get_dummies(df_clean['education'],prefix='Education').astype(int)
employ=pd.get_dummies(df_clean['self_employed'],prefix='Employed').astype(int)

```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
le=LabelEncoder()
df_clean['Loan_Status_Encoded']=le.fit_transform(df_clean['loan_status'])
```

```python
df_final=pd.concat([employ,edu,df_clean],axis=1)
```

```python
df_final.drop(['education','self_employed','loan_status'],axis=1,inplace=True)
```

```python
df_final
```

```python
Scaler=MinMaxScaler()
df_scaled = pd.DataFrame(Scaler.fit_transform(df_final), columns=df_final.columns, index=df_final.index)
```

```python
x=df_scaled.drop(['loan_id','no_of_dependents','Loan_Status_Encoded'],axis=1)
```

```python
y=df_scaled['Loan_Status_Encoded']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain,xtest,ytrain,ytest =train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=SVC()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
ypredicted=model.predict(xtest)
```

```python
from sklearn.metrics import f1_score,precision_score,recall_score
```

```python
precision = precision_score(ytest, ypredicted)
recall = recall_score(ytest, ypredicted)
f1 = f1_score(ytest, ypredicted)
```

```python
print(f'Precision is',precision)
print(f'Recall Score is',recall)
print(f'f1 score is',f1)
```
